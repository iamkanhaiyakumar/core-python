SERIES

Series:-
A pandas Series is a oneâ€‘dimensional labeled array that can hold any data type. It pairs each value with an index label, enabling fast, vectorized operations.

Key points:
- 1D: single column of data with an index
- Heterogeneous types allowed
- Index labels enable intuitive selection (e.g., var['c'], var[2])

Examples already used:
- From list with custom index, dtype, and name:
    pd.Series(x, index=['a','b','c','d','e'], dtype='float', name='My Series')
- From dictionary (keys become index, values become data):
    pd.Series(dic)

Access:
- Positional: var[2]
- Label: var['c']

```python
%pip install pandas
import pandas as pd
```

```python
x=[10,20,30,40,50] # series from list


var=pd.Series(x) # create series
var=pd.Series(x,index=['a','b','c','d','e']) # create series with custom index
var=pd.Series(x,index=['a','b','c','d','e'],dtype="float") # create series with custom index
var=pd.Series(x,index=['a','b','c','d','e'],dtype="float",name='My code') # create series with name

print(var) # print series
print(type(var)) # print type of series
print(var[2]) # print value at index 2
```

```python
dic = {"name":['python','java','c++'],'rank':[1,2,3],'year-of-release':[1991,1995,1985]} # series from dictionary

var1= pd.Series(dic) # create series from dictionary

print(var1) # print series from dictionary
```
