Arithmetic operations

Pandas allows element-wise arithmetic on Series and DataFrames.
Operations are aligned by index or column names.

1. Arithmetic on Series

```python
import pandas as pd

s1 = pd.Series([10, 20, 30])
s2 = pd.Series([1, 2, 3])

print(s1 + s2)   # Addition
print(s1 - s2)   # Subtraction
print(s1 * s2)   # Multiplication
print(s1 / s2)   # Division
```

2. Arithmetic on DataFrame

```python
df1 = pd.DataFrame({'A': [10, 20], 'B': [30, 40]})
df2 = pd.DataFrame({'A': [1, 2],  'B': [3, 4]})

print(df1 + df2)
print(df1 - df2)
print(df1 * df2)
print(df1 / df2)
```

```python
df3 = pd.DataFrame()
df3["C"]= df1["A"] + df1["B"]
print(df3)
```

```python
var = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
print(var)
```

```python
var["C"] = var["A"] + var["B"]
print(var)
```

```python
var["C"] = var["A"] - var["B"]
print(var)
```

```python
var["C"] = var["A"] * var["B"]
print(var)
```
