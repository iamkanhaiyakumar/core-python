DataFrame:-

Definition

DataFrame: A pandas DataFrame is a twoâ€‘dimensional, labeled tabular data structure (rows and columns) for Python that can hold heterogeneous data types. It is built on NumPy and provides fast, flexible operations for data cleaning, transformation, aggregation, and I/O.
Key features

2D table: Rows (index) and named columns.
Heterogeneous types: Different columns can have different dtypes.
Indexing: Label- and position-based selection (.loc, .iloc).
Alignment: Automatic alignment on operations by index/column labels.
Missing data: Built-in handling (e.g., NaN, .fillna(), .dropna()).
Grouping & aggregation: groupby, agg, transform.
Joins/merges: merge, join, concat.
I/O: Read/write CSV, Excel, SQL, JSON, Parquet, etc.
Minimal example

Create a DataFrame from a dictionary:
import pandas as pd
df = pd.DataFrame({'name': ['Alice','Bob'], 'age': [30, 25]})
print(df)
